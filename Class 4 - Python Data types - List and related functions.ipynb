# Data type:

*In Python the entire data types can be classified into 2 types. One is Primitive and another one Non Primitive.
 


* Primitive Datatype: 
  * A data type that can't be broken down into a more simple data type.
  * In simple terms these data types which are pre-defined and supported by the programming language.
  * Ex: Integer, float, boolean, String.
* Non Primitive Datatype: 
  * These data types which are derived from Primitive data types and offer increased functionality.
  * In simple terms these data types just don't store the value, but rather a collection of values in various formats.
  * Ex: Lists, Tuples, Dictionary, Sets.

# List 

* In simple terms it is a collection of data elements which is stored in a single variable.
* It's a collection of data elements related to homogeneous or hetrogeneous data. 
* It represents an ordered, mutable collection of objects.  
* It can be considered as the most general version of a 'Sequence". Unlike the strings they are mutable which means the elements inside a list can be changed. 
* Lists are constructed with brackets[] and commas separating every element in the list.
* We can access the elements of the list by using Index.
    * Ex: Lists are collection of items[Strings,Integers or even other lists].

```python
# Creating a list using open and closed squared braces.

List1 = []  # Creating an empty list.
print(List1)
print(type(List1))

# Creating a list of integers - Homogeneous datatype.
List2 = [10,20,30,40,50]
print(List2)

# Creating a list of strings - Homogeneous datatype.
List3 = ["Sainath","Business Analyst","L&T Infotech"]
print(List3)

# Creating a list of different data types  - Non Homogeneous datatype.
List4 = ["Sainath",50000,5.8,75,"True"]
print(List4)

# Creating a list with another list - List of lists
Emp_details = [List3,List4]
print(Emp_details)
```

```python
# Creating a list using list function - list()
# If no argument is given, then it creates a new emplty list.
# We need to pass only one argument.
# The argument which we are passing it must be "iterable". Iterable means either string or dictionary or tuples or set or lists.

L1 = list()
print(L1)
print(type(L1))


L2 = ["Sainath","Business Analyst", "Legato Healthe Care Technologies", "Bangalore"]
print(L2)
print(type(L2))


L3 = list("Python")
print(L3)
print(type(L3))


L4 = list("1986")
print(L4)
print(type(L4))


L5 = list(L2)
print(L5)
print(type(L5))


# The main difference between [] and list() function is that 
# When we are creating a list uisng [] then we can pass any datatype of homogeneous or heterogeneous.
# When we are creating a list using list() we need pass iterable objects only.
```

# List Indexing & Slicing

* We can access individual elements from the list is called as 'Indexing'.
* We can access a range of element from the list is known as slicing.
* In the list the index always starts from 0 i.e., from left to right.
* If we want to access the elements from right side of list we can use "Negative" index. Here the index starts from -1,-2 etc.,

```python
# Example : 1 - Indexing

List5 = [10,20,30,40,50]

# Print 2nd element from the list.
print(List5[1])
print(type(List5[1]))

# Print the last element from the list.
print(List5[-1])

# Example : 2 - Slicing

#Print the entire list
print(List5[:])

# Print the element in the list from 1st index to the last
print(List5[1:])

# Print the element from 0th index to 3rd index
print(List5[:3]) # Print the list of element upto before the end position


# Print the entire list in reverse direction
print(List5[::-1])


# print the entire list in reverse direction with stepsize 2
print(List5[::-2])


# Print the list from right to left from -1 to -3rd index
print(List5[-3:-1])
```

```python
# Creating a list using another list.

L6 = ["Sainath","Business Analyst","IT Sector"]
L7 = [1615515,5.8,1986,90000 ]

L8 = [L6,L7]
print(L8)
print(type(L8))

# Accessing the elements from the list

print(L8[0])

print(L8[1])

# Accessing the element with in the list from the list.

print(L8[1][3])  # First argument is related to index in the list and the second argument is related to index in the index.

print(L8[-1][-3])

print(L8[-1][-3:-1])

print(L8[1][0:3])

print(L8[0:2])

print(L8[0:10])  # Whenever we are slicing the list using range, even if we pass the out of the index, then we don't get any error.
# Simply it will print the entire list.

print(L8[2])  # We will get an error message like "Index Out of range"
```

```python
# Another Example:

L9 = ["Sainath","030163", 1986, 5.8, 90000,"Business Analyst",["AP","Krishna","Vijayawada",520013] ]
L9

print(L9[6])

print(L9[6][-1])      # Printing the right side of the element in the list with in the list.

print(L9[: : -1])     # Printing the entire list in reverse direction.

print(L9[6][: : -1])  # Printing the list with in the list in reverse direction.

print(L9[0:7])  # Printing the list from oth Index to 6th Index.

print(L9[-1:-6])  #  No Output. Due to -ve slicing and with positive stepsize "1 (Default size)".

print(L9[-1:-50:-1])

print(L9[0:50])  

print(L9[7])  # When we are slicing the exact index which is out of range then we will get "Index Error" message.
```

# List Methods

```python
# Concatenation -  We can combine two lists by using "+" symbol.

List1 = ["Sainath",35,5.8,600000]
List2 = ["L&T Infotech","Hyderabad"]

Emp_details = List1+List2
print(Emp_details)


# Length - We can find the total number of elements present in the list.
print(len(Emp_details))


# Repetition - If we want to print the list for required number of times by uinsg "*" symbol.
print(List1*5)

# Membership - To check whether the value is presented in the list or not by using "in" operator.

# Using "in" operator
print("Sainath" in List1)

# Using "not in" operator
print(600000 not in List1)


# To find the Smallest number in the list
List3 = [10,1.5,.3,10.6,59,0.01,100,65,10]
print("The smallest number in the give list is", min(List3))


# To find the largest number in the list
List3 = [10,1.5,.3,10.6,59,0.01,100,65,10]
print("The largest number in the give list is", max(List3))


# To find the sum of all elements in the list
print("The sum of all elements in the list is",sum(List3))


# Append - It will the required element in the end of the list and return none.
List4 = ["Sainath",10666246,"L&T Infotech"]
List4.append("Hyderabad")
print(List4)

List5 = ["Hyderabad","Skyview","P&G,e-Pragati"]
List5.append(10666246)
print(List5)
List5.append(List4)
print(List5)

# Note: 
# If we concatenate i.e., adding then this will happen for new list where as append will add the element in the same list.

# Extend - Appends all the elements from one list to the another list and returns none.
List4 = ["Sainath",10666246,"L&T Infotech"]
List5 = ["Hyderabad","Skyview","P&G,e-Pragati"]
List4.extend(List5)
print(List4)

# Note : The difference between "append and extend" is that when we add a new list to another list, then append will add the new list as list.
# Where as extend will add the list as elements to another list.
# Output - append: ['Hyderabad', 'Skyview', 'P&G,e-Pragati', 10666246, ['Sainath', 10666246, 'L&T Infotech', 'Hyderabad']]
# output - Extend: ['Sainath', 10666246, 'L&T Infotech', 'Hyderabad', 'Skyview', 'P&G,e-Pragati']

# Insert - It used to insert the data element at the required position in the list.
List5.insert(0,10666246)  # The first argument is the position and next argument is the required data element.
print(List5)

List5.insert(-2,"Business Analyst") # Incase if we use -ve index, then it replaces the element before the position of the mentioned index.
print(List5)

# Difference between append and insert function:
# Append is used to place the element at the end of the list, where as insert function will place the element at the required position.
# Incase of insert function, if we use +ve index then it shifts the elements at the required position and add the element.
# Incase of -ve index insert function will add the element before the index in the list.
```

# POP Function:

Used POP to "POP Off (Remove)" an item from the list and returns it. By default POP takes the last index. In case if we want specify a particular index we use the required index and display the element.

# Remove Function: 

It remove the required data element from the list which is occured first.

# Difference between POP and Remove 
* Both removes the data element. 
* POP function removes the element based on the index.
* Remove function deletes the element according to mentioned data element or value.

```python
L = [1,2,3,4,"Sainath","Busienss Analyst"]

L1 = L.pop() # By default it takes the last index and remove the element.

L2 = L.pop(0)  # Remove the element at the oth index.

L3 = L.pop(-1) # Remove the element at the mentioned index.

print(L1)
print(L2)
print(L3)

L1 = [1,2,2,1,3,4,5,1,2,9]
L1.remove(2)  # Removes the required element which is occured for first occurnace.
print(L1)

# Note: By using "del" keyword we can remove the data element only when we know the exact index but in remove we can delete the element directly without knowing the index.

# Count - Returns the number of times element appears in the list.
print(L1.count(2))


# Reverse - Used to print the list in reverse direction. The changes will be done in the list permenantly.
L1.reverse()
print(L1)

# Another way of printing
L1[::-1]  # Changes will not be done in the list. Changes applicable only for print purpose.
L1
```

```python
# Sort - Used to sort the elements of a given list in a specific order - Ascending(Default) or Decending. 
# The original list is not changed.

Vowels = ["e","a","u","i","o",]

# Sorting in ascending Order
Vowels.sort()
print("After sorting in ascending order of the elements are", Vowels) 

# Sorting in decending Order
Vowels.sort(reverse = True)
print("After sorting in decending order of the elements are", Vowels)


Num = [1,99,10,45,89,36,74,100]

# Sorint in ascending order
Num.sort()
print("After sorting in ascending order of the elements are",Num)

# Sorting in decending order
Num.sort(reverse=True)
print("After sorting in decending order of the elements are",Num)
```

# Range () Function - It will generate a data set.
* We can generate a sequence of numbers using range() function. 
* If we write range(10) python will generate numbers from 0 to 9. 
* We need to define the starting position, ending position and the increment value i.e.,start value, end value and step size  as range. By default step size is "1" if not provided.

```python
print(range(10))  # It will generate a object.

print(list(range(10)))

print(list(range(1,35,2)))   # It will generate data from the starting point 1 to ending point 35 with increment by "2".

print(list(range(20,2,-2)))

print(list(range(5,45,5)))
```

```python
print(list(range(10,1)))  # No output. Due based on the scaling.As we moving 10,9,8,7,...i.e., decreasing with default size "1(+ve)"

print(list(range(10,1,-1))) # Output

print(list(range(1,10,-1)))  # No output.

print(list(range(1,10.5)))  # In range function we should use only integer values. No floating values are allowed.
```

# List Comprehensions

* It's an elegent and concise way to create new list from an existing list in python.
* It's consists of an expression followed by for statement inside square brackets.


```python
# Example:1  Add the cube of first 10 numbers to the empty list

# Without list comprehension
Cube = []

for i in range(10):
    Cube.append(i**3)
print(Cube)


# With comprehension
Cube1 = [i**3 for i in range(10)]
print(Cube1)
```

```python
# Example:2  Adding a data elements to a list by using for loop.

# Without list comprehension
L = []

for i in ["Sainath","Business Analyst","L&T Infotech"]:
    L.append(i)

print(L)

# With list comprehension

L1 = [i for i in ["Sainath","Business Analyst","L&T Infotech"]]
print(L1)
```

```python
# Print the even and odd numbers between 0 and 20.

Even = [x for x in range(20) if x%2 == 0]
print("The even numbers in the list are", Even)

Odd = [x for x in range(20) if x%2 != 0]
print("The odd numbers in the list are",Odd)
```

```python
# Example:3  Program to find the sum of all number in a list.

# Without list comprehension
List = [10,25,15,89,91,105,3,2]

sum = 0

for i in List:
    sum = sum+i
    
print("The sum of numbers in the list is",sum)
```

# Nested List:

It means we can have data structures within data structures.It's one of the great features in Python data structures.

```python
List_1 = [1,2,3]
List_2 = [4,5,6]
List_3 = [7,8,9]

# Make a list of lists in the form of matrix.

Matrix = [List_1,List_2,List_3]

print(Matrix)
 
[[1, 2, 3], [4, 5, 6], [7, 8, 9]]  # In the output [1,2,3] has the index "0", [4,5,6] has the index "1" and [7,8,9] has the index 2.


# Accessing the elements from the matrix.

print(Matrix[0])

print(Matrix[1])

print(Matrix[2])

Matrix[0][2]  # Accessing the 2nd data element from 0th index.

Matrix[2][2]


type(Matrix)
```

```python
# Extract the 1st index of each element from the matrix and prepare a list.

List_1 = [1,2,3]
List_2 = [4,5,6]
List_3 = [7,8,9]

# Make a list of lists in the form of matrix.

Matrix = [List_1,List_2,List_3]
# Step-1 : First we initiate an empty list.

l = []

# Step-2: We create a loop function to access the required element.

for i in Matrix:
    l.append(i[1])

# Explanation - Here i takes the values like 0,1,2 in for loop i.e., those are nothing but indexes of the list in the matrix.
# Then we need to capture the 1st element in each list. That's why we mention the index 1 in append function.
    
print(l)

M = []

for i in Matrix:
    M.append(i[2])
    
print(M)


# Another build function which is used to prepare the same list  without using list function.

First_col = [row[1] for row in Matrix]  # This function will create list by capturing the 1st index in each row of the index.
print(First_col)

# Create a list with 2nd element from the matrix.

Second_col = [row[2] for row in Matrix]
print(Second_col)


# We can prepare a list with square of 1st element from the matrix.

Square_col = [row[1]**2 for row in Matrix]
print(Square_col)

# Prepare a list with two elements.

Two_List = [(i[0],i[1]) for i in Matrix]

print(Two_List)

```
