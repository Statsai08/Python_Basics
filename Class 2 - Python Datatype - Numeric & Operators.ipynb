# COMPARISION & RELATIONAL OPERATORS

Comparision Operators are used to compare the values. It either returns TRUE Or FALSE based on the condition.
The comparision opertors are also known as "Relational Operators".


```python
# Sample Code

a = 20
b = 30

print(type(a>b))    # For comparision operator the result is  always boolean.

print('{} > {} is'.format (a,b),a>b) # Here output will be formatted. During the execution, the value of a and b be will taken
                                    # into curly braces and then condition will be checked.
    
# Greater than condition
print('{} > {} is'.format (a,b), a>b)

# Less than condition
print('{} < {} is'.format(a,b), a<b)

# Equal to condition
print('{} =={} is'.format(a,b), a==b)

# Not equal to condition
print('{} != {} is'.format(a,b), a!=b)

# Greaterthan or equal to
print('{} >= {} is'.format(a,b), a>=b)

#Lessthan or equal to
print('{} <= {} is'.format(a,b), a<=b)
```

# BITWISE OPERATORS

* Bitwise operators are used to compare (binnary) numbers.It works on bits and performs bit by bit operation.
* First we have to convert the decimal numbers to the binary numbers and then do the comparison




# Method to convert a number to binary number

syntax : '{0:08b}'.format(number)

Where,
    {} - place a variable into a sting.
    0  - takes the variable at the argument position 0.
    :  - adds formating options for this variable. Otherwise it would represent decimal 6.
    08 - format the number to 8 digits zero-padded in the left.
    b  - converts the number into it's binary representation


```python
a = 17
b = 10

print('Binary number for {0} is {0:08b}'.format(a))
print('Binary number for {0} is {0:08b}'.format(b))
```

```python
# Binary "And" operator  - Sets each bit to 1 if both bits are 1 returns 1 otherwise returns 0. 
# If we compare the binary numbers for each digit for both a and c the result binary will be 00000000

print('Binary AND value is', a&b)       

# Binary "OR" opertor  - Sets each bit to 1 if one of the bits are 1 returns 1 otherwise returns 0.
# If we compare the binary numbers for each digit for both a and c the result binary will be 00011011
print('Binary OR value is', a|b)

# Binary "XOR" operator - Sets each bit to 1 if only one of two bits is 1.
# If we compare the binary numbers for each digit for both a and c the result binary will be 00011011
print('Binary XOR value is',a^b)

# Binary "Not" operator - Sets to all the ones to zero and and all the zero's to ones. It inverts all the bytes.
# If we compare the binary numbers for each digit for the value "a", the result binary will be 11101110
print('Binary Not value is',~a)

# If we compare the binary numbers for each digit for the value "b", the result binary will be 11110101
print('Binary Not value is',~b)


# Binary right shift operator - Sets the bit i.e., shift right by pushing copies of the left most bit in from the left, and let the rightmost bits fall off
# Based on the vlaue mentioned, it will shift the position to the right and truncate the last digits mentioned.
# i.e., it will add "Zeros" according to the number to the left side of binary digit and ignore the same number of digits rightside of the binary digit.
# Ex: For the value "a" the binary code is "00010001" if use right shift with 2 then it will add two zeros before and ignore last two digits at the end.
# Then the new binary code is "00000100"
print('Binary Right Shift is', a>>2)


# Similarlly we can use left shift.
print("Binary Left Shift is", a<<2)

```

# SPECIAL OPERATORS
Python language supports two types of special operators. 1: Identity Operators 2: Membership Operators.

# 1: Identity Operator

* These operators are used to compare the memory location of two objects.
* Is and Isnot are two identity operators in python.
* Output will be a boolean value.


```python
# "Is" Operator returns "True" if the memory address is same
# "Is Not" operator returns "True" if the memory address is different
```

```python
# Example - 1  Numbers using  "Is" operator

x = 10
y = 10

p =1
q =2


print('Memory location of x is',id(x))   # Id operator is used for find the memory location of variable.
print('Memory location of y is',id(y))
print (x is y)  # Comparing the memory location of two variables.
print (x is not y )  # If both values are different then it returns "True". If both values are same it returns "False"
print (p is q)
print (p is not q)
```

```python
# Example - 2 String using "Is" operator

a = 'Python'
b = 'Python'

print('Memory location of a is',id(a))
print("Memory location of b is", id(b))
print( a is b)
print( a is not b)
```

```python
# Example - 3 List using "Is" operator

list1 = [1,2,3,4]
list2 = [1,2,3,4]

print("Memory location of list1 is",id(list1))
print("Memory location of list2 is",id(list2))
print(list1 is list2)
print(list1 is not list2)
```

# MEMBERSHIP OPERATOR

* Membership operators are used to test whether a value or variable is found in a sequence(Eg: String, list, tuple, set and dictionary)
* "In" and "Not in" operators are membership operators in python.


```python
# "In" operator will display "True" if the value is found in the sequence.
# "Not In" operator will display "True" if the value is not found in the sequence i.e., exactly opposite.
```

```python
# Example -1

mylist = [1,1.5,'python',20]
print(1 in mylist)
print("python" in mylist)
print(30 in mylist)
```

```python
# Example -2

mylist1 = [10,20.4,'JAVA',45]
print(50 not in mylist1)
print('JAVA' not in mylist1)
```
