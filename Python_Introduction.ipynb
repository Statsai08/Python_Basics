# What is Python?
* Python is a High-level, interactive, interpreted and object-oriented.
* Python is High Level - Written in a form that is close to our human language. No particular knowledge of the hardware is needed as high level languages create programs that are portable and not tied to a particular computer or microchip.
* Python is Interactive – Programmer can interact with the interpreter directly to write your programs using Python prompt.
* Python is Interpreted − Python is processed at runtime by the interpreter. You do not need to compile your program before executing it.
* Python is Object-Oriented − Python supports all the Object-Oriented techniques such as Class, Object, Methods, Inheritance ,Encapsulation and Polymorphism

# The difference between Complier and Interpreter:
* Compiler transform the code written in a high-level program statement into machine code at once before program runs.
* Interpreter converts each high-level program statement into machine code one by one during the program runs.
* Note: In simple terms ""Compiler compiles the code in a complete chunk and an Interpreter compiles a code line by line and convert the code into machine language.""

# Jupyter Notebook Coding: 
* To create a new code cell just press “a” in the keyboard it will create a new line to the above code.
* To create a new code cell just press “b” in the keyboard it will create a new line to the below code.
* To create comment in the coding we use “#”. Whatever we write after the # then system will not be executed by interpreter (A program that reads and executes code). We have two types of command.
    * Inline command – Providing comments in the code which will help the user to understand about the code.
    * Block command – Providing multiple line comments in the code which will help the user to understand about the code.
* To enter any comments for the code, we can select “Markdown” to enter the comments.
* Green Color indicates the cell is in "Edit" mode. To close edit mode "Press Escape KeY".
* Blue color indicates "Commond Mode".
* Execution can be done in 3 ways.  
    * Ctrl + Enter (Run the code and will not create a new cell)
    * Shift + Enter (Run the code and create a new cell)
    * Run button on jupyter interface.
* Important Shortcut keys: These keys will be applicable when the cell in "Commond Mode" i.e., Blue color.
    * a --> to create a cell above the active cell
    * b --> to create a cell below the active cell
    * d + d --> for deleting cell
    * m --> for markdown cell
    * v --> for code cell
